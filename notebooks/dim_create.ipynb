```python
import pandas as pd

# 1. Ustaw zakres dat (np. 10 lat wstecz i 5 do przodu)
start_date = "2000-01-01"
end_date = "2030-12-31"

# 2. Generowanie zakresu
dates = pd.date_range(start=start_date, end=end_date)

# 3. Tworzenie DataFrame
df = pd.DataFrame(dates, columns=['full_date'])

# 4. Wyliczanie kolumn
df['date_key'] = df['full_date'].dt.strftime('%Y%m%d').astype(int)
df['date_str'] = df['full_date'].dt.strftime('%Y-%m-%d') # Klucz dla NiFi
df['year'] = df['full_date'].dt.year
df['month'] = df['full_date'].dt.month
df['day'] = df['full_date'].dt.day
df['day_name'] = df['full_date'].dt.day_name()
# Weekend: 5=Sobota, 6=Niedziela
df['is_weekend'] = df['full_date'].dt.dayofweek.isin([5, 6]) 

# drop first column
df = df.drop(columns=['full_date'])

# 5. Zapis do CSV (bez nagłówka, gotowy do importu SQL)
df.to_csv('dim_date_load.csv', index=False, header=True)

print("Plik dim_date_load.csv został utworzony. Zaimportuj go do bazy.")
```
